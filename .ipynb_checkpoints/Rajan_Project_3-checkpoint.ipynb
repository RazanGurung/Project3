## Data Cleaning

```python
missing_counts=df.isnull().sum()
missing_counts
```

```python
# Calculate missing values
missing_counts = df.isnull().sum()
features_withNaN = missing_counts[missing_counts > 0]

if not features_withNaN.empty:
    missing_data = pd.DataFrame({
        'Missing Count': features_withNaN,
        'Percentage (%)': (features_withNaN / len(df)) * 100
    }).sort_values(by='Missing Count', ascending=False)
    
    missing_data['Percentage (%)'] = missing_data['Percentage (%)'].round(2)
    
    print("Features with Missing Values:")
    print(missing_data)
else:
    print("No features with NaN values found.")
```

```python
# Drop columns with too many missing values more than 60 percent
df.drop(['Pool QC', 'Misc Feature', 'Alley', 'Fence'], axis=1, inplace=True)

# Replace FireplaceQu NaN with "No Fireplace"
df['Fireplace Qu'].fillna('No Fireplace', inplace=True)
```

The missing data in 'Pool QC', 'Misc Feature', 'Alley', 'Fence' are too high and altering them to use would not help the model but create noise and encourage model biasness so we prefered to drop them.
'

```python
garage_cols = ['Garage Type', 'Garage Finish', 'Garage Qual', 'Garage Cond']
df[garage_cols] = df[garage_cols].fillna('No Garage')

# For Garage Yr Blt, impute with Year Built (since no garage = no year)
df['Garage Yr Blt'].fillna(df['Year Built'], inplace=True)
```

```python
bsmt_cat_cols = ['Bsmt Qual', 'Bsmt Cond', 'Bsmt Exposure', 'BsmtFin Type 1', 'BsmtFin Type 2']
df[bsmt_cat_cols] = df[bsmt_cat_cols].fillna('No Basement')
```

```python
df['Lot Frontage'] = df.groupby('Neighborhood')['Lot Frontage'].transform(
    lambda x: x.fillna(x.median())
)
```

```python
df['Mas Vnr Type'].fillna('None', inplace=True)
df['Mas Vnr Area'].fillna(0, inplace=True)
```

```python
df['Electrical'].fillna(df['Electrical'].mode()[0], inplace=True)
```

```python
bsmt_num_cols = ['BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF', 'Total Bsmt SF', 'Bsmt Full Bath', 'Bsmt Half Bath']
df[bsmt_num_cols] = df[bsmt_num_cols].fillna(0)
```

In the above cells, the missing columns are handled in various ways as per the need of the features itself. The numerical data are filled with its median and the categorical variables are handled with the neutral values.

### Feature Engineering

```python
#Buyers care about total livable space, not just above-ground
df['TotalSF'] = df['Gr Liv Area'] + df['Total Bsmt SF']
```

```python
df.columns
```

```python
#summing all the bathrooms in the house
df['TotalBath'] = df['Full Bath'] + (0.5 * df['Half Bath']) + \
                  df['Bsmt Full Bath'] + (0.5 * df['Bsmt Half Bath'])
```

```python
#Age of the house and categoring into new or remodel
df['Age'] = df['Yr Sold'] - df['Year Built']
df['Remodeled'] = (df['Year Remod/Add'] != df['Year Built']).astype(int)
```

```python
#Total porch space
df['TotalPorchSF'] = df['Open Porch SF'] + df['Enclosed Porch'] + \
                     df['3Ssn Porch'] + df['Screen Porch']
```

```python
#if has basement or not
df['HasBasement'] = df['Total Bsmt SF'].apply(lambda x: 1 if x > 0 else 0)
```

```python
#if house has garage or not
df['HasGarage'] = df['Garage Area'].apply(lambda x: 1 if x > 0 else 0)
```

```python
#if the house has fireplace or not
df['HasFireplace'] = df['Fireplaces'].apply(lambda x: 1 if x > 0 else 0)
```

```python
df["Neighborhood"].unique()
```

```python
#neighborhood which are less relavent are grouped as others
neighborhood_counts = df['Neighborhood'].value_counts()
rare_neighborhoods = neighborhood_counts[neighborhood_counts < 10].index
df['Neighborhood_Grouped'] = df['Neighborhood'].replace(rare_neighborhoods, 'Other')
```

```python
df['Neighborhood_Grouped'].unique()
```

```python
#grouping house by style type 
df['HouseStyle_Simple'] = df['House Style'].replace({
    '1.5Unf': 'MultiLevel', '1.5Fin': 'MultiLevel',
    '2.5Unf': 'MultiLevel', '2.5Fin': 'MultiLevel',
    'SFoyer': 'SingleStory', 'SLvl': 'SingleStory',
    '1Story': 'SingleStory', '2Story': 'MultiLevel'
})
```

```python
df['HouseStyle_Simple'].unique()
```

```python
#High-quality materials in a large home may boost value
df['Qual_x_Area'] = df['Overall Qual'] * df['Gr Liv Area']
```

```python
#bathrooms per room
df['BathPerBedroom'] = df['TotalBath'] / df['Bedroom AbvGr']
df['BathPerBedroom'].replace(np.inf, 0, inplace=True)  
```

```python
df['Mo Sold'].unique()
```

```python
#converting months to quater
df['QuarterSold'] = df['Mo Sold'].replace({
    1: 1, 2: 1, 3: 1,      
    4: 2, 5: 2, 6: 2,    
    7: 3, 8: 3, 9: 3,     
    10: 4, 11: 4, 12: 4    
})
```

```python
#time since renovation
df['YearsSinceRemodel'] = df['Yr Sold'] - df['Year Remod/Add']
```

```python
#price per square foot
df['PricePerSF'] = df['SalePrice'] / df['TotalSF']
```

```python
df['Lot Shape'].unique()
```

```python
#numeric score for lot shape
df['LotShape_Score'] = df['Lot Shape'].map({
    'Reg': 3,   
    'IR1': 2,   
    'IR2': 1,  
    'IR3': 0    
})
```

Various new features were created using two or more features where the initial feature are combined with the redundant features or features providing very close information so that the dimensionality of the dataframe is reduced without loosing the key information.

```python
# Set up the plotting function
def plot_features(plot_type, features, df):
    n_cols = 3
    n_rows = (len(features) + n_cols - 1) // n_cols
    fig, axes = plt.subplots(n_rows, n_cols, figsize=(18, 5*n_rows))
    axes = axes.flatten()
    
    for i, col in enumerate(features):
        ax = axes[i]
        try:
            if plot_type == 'hist':
                sns.histplot(data=df, x=col, kde=True, ax=ax)
                ax.set_title(f'Distribution of {col}')
            elif plot_type == 'count':
                sns.countplot(data=df, x=col, ax=ax)
                ax.set_title(f'Count of {col}')
                ax.tick_params(axis='x', rotation=45)
        except:
            print(f"Could not plot {col}")
            continue
        
    # Hide empty subplots
    for j in range(i+1, len(axes)):
        axes[j].set_visible(False)
    
    plt.tight_layout()
    plt.show()

# Plot histplots for continuous variables
print("Continuous Variable Distributions:")
plot_features('hist', histplot_features, df)

# Plot countplots for discrete variables
print("\nDiscrete Variable Counts:")
plot_features('count', countplot_features, df)

# Plot countplots for categorical variables
print("\nDiscrete Variable Counts:")
plot_features('count', categorical_features, df)
```

### Feature Selection

```python
engineered_features = [
    'TotalSF', 'TotalBath', 'Qual_x_Area', 
    'Age', 'YearsSinceRemodel', 'TotalPorchSF','HasBasement', 'HasBasement',
    'HasGarage', 'HasFireplace', 'BathPerBedroom', 'QuarterSold', 'PricePerSF', 'LotShape_Score'
]
numeric_features = [f for f in numeric_cols if f not in engineered_features + ['SalePrice', 'Order', 'PID']]
```

```python
numeric_features
```

```python
# Calculate correlation with target
corr_with_target = df[numeric_features + ['SalePrice']].corr()['SalePrice'].abs().sort_values(ascending=False)

# Select features with correlation > threshold
strong_corr_features = corr_with_target[corr_with_target > 0.4].index.tolist()
strong_corr_features.remove('SalePrice')
```

To identify the most influential features for predicting SalePrice, we performed correlation-based feature selection. We calculated the absolute Pearson correlation between each numerical feature and the target, retaining only those with a correlation > 0.4 to ensure meaningful linear relationships. This efficient filter method reduces dimensionality by eliminating weak predictors while maintaining interpretability. However, since it only captures linear dependencies, we may later explore non-linear relationships and feature interactions for further refinement. The selected features strong_corr_features will be used in subsequent modeling stages.

```python
strong_corr_features
```

```python
# ANOVA test for categorical features
from sklearn.feature_selection import f_regression

categorical_selected = []
for cat_feature in categorical_features:
    dummies = pd.get_dummies(df[cat_feature], drop_first=True)
    f_scores, _ = f_regression(dummies, df['SalePrice'])
    if max(f_scores) > 10:  # Threshold for significance
        categorical_selected.append(cat_feature)
```

We performed ANOVA F-tests to identify significant categorical features affecting SalePrice. Each categorical variable was one-hot encoded, and we computed F-scores against the target variable. Features with at least one category showing an F-score >10 were selected for modeling. This method efficiently identifies meaningful categorical predictors while maintaining statistical rigor. The selected features will be combined with our numerical features in subsequent modeling stages, with potential follow-up analysis to address high-cardinality features or non-linear relationships.

```python
categorical_selected
```

```python
candidate_features = strong_corr_features + categorical_selected + engineered_features + ['SalePrice']

# Remove redundant features
redundant_pairs = {
    'TotalSF': ['Gr Liv Area', 'Total Bsmt SF','1st Flr SF'],
    'TotalBath': ['Full Bath', 'Half Bath', 'Bsmt Full Bath', 'Bsmt Half Bath'],
    'TotalPorchSF': ['Open Porch SF', 'Enclosed Porch', '3Ssn Porch', 'Screen Porch'],
    'HasFireplace': ['Fireplaces'],
    'Neighborhood_Grouped':['Neighborhood'],
    'LotShape_Score':['Lot Shape'],
    'HouseStyle_Simple':['House Style'],
    'HasGarage':['Garage Area', 'Garage Cars', 'Garage Yr Blt','Garage Type','Garage Finish', 'Garage Qual','Garage Cond'],
    'HasFireplace':['Fireplaces', 'Fireplace Qu'],
    'Age':['Yr Sold','Year Built' ],
    'Remodeled':['Year Remod/Add']
}

features_to_remove = []
for main_feature, redundant in redundant_pairs.items():
    if main_feature in candidate_features:
        features_to_remove.extend(redundant)

final_features = [f for f in candidate_features if f not in features_to_remove]
```

```python
from sklearn.preprocessing import StandardScaler, OneHotEncoder, OrdinalEncoder
from sklearn.compose import ColumnTransformer
from sklearn.pipeline import Pipeline

# Define the order for ordinal features
ord_categories = [
    ['Po', 'Fa', 'TA', 'Gd', 'Ex'],  # Exter Qual
    ['Po', 'Fa', 'TA', 'Gd', 'Ex'],  # Exter Cond
    ['No Basement', 'Po', 'Fa', 'TA', 'Gd', 'Ex'],  # Bsmt Qual
    ['No Basement', 'Po', 'Fa', 'TA', 'Gd', 'Ex'],  # Bsmt Cond
    ['Po', 'Fa', 'TA', 'Gd', 'Ex'],  # Kitchen Qual
    ['Po', 'Fa', 'TA', 'Gd', 'Ex'],  # Heating QC
    ['Sal', 'Sev', 'Maj2', 'Maj1', 'Mod', 'Min2', 'Min1', 'Typ'],  # Functional
    ['N', 'P', 'Y'],  # Paved Drive
    ['No Basement', 'No', 'Mn', 'Av', 'Gd'],  # Bsmt Exposure
    ['No Basement', 'Unf', 'LwQ', 'Rec', 'BLQ', 'ALQ', 'GLQ'],  # BsmtFin Type 1
    ['No Basement', 'Unf', 'LwQ', 'Rec', 'BLQ', 'ALQ', 'GLQ']  # BsmtFin Type 2
]
# Create transformers
num_transformer = StandardScaler()
cat_transformer = OneHotEncoder(handle_unknown='ignore', sparse_output=False)
ord_transformer = OrdinalEncoder(categories=ord_categories)
binary_transformer = 'passthrough'  

# Combine all transformers
preprocessor = ColumnTransformer(
    transformers=[
        ('num', num_transformer, num_features),
        ('cat', cat_transformer, cat_features),
        ('ord', ord_transformer, ord_features),
        ('binary', binary_transformer, binary_features)
    ],
    remainder='drop'  # Drop any columns not specified
)

# Create a processing pipeline
pipeline = Pipeline(steps=[
    ('preprocessor', preprocessor)
])

# Apply preprocessing
processed_data = pipeline.fit_transform(df)

# Get feature names for the processed data
# For numerical features (kept original names)
num_feature_names = num_features

# For one-hot encoded features
cat_encoder = pipeline.named_steps['preprocessor'].named_transformers_['cat']
cat_feature_names = cat_encoder.get_feature_names_out(cat_features)

# For ordinal features (kept original names)
ord_feature_names = ord_features

# For binary features (kept original names)
binary_feature_names = binary_features

# Combine all feature names
all_feature_names = list(num_feature_names) + list(cat_feature_names) + \
                   list(ord_feature_names) + list(binary_feature_names)

# Convert to DataFrame for better readability
processed_df = pd.DataFrame(processed_data, columns=all_feature_names)
```

```python
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import GradientBoostingRegressor

models = {
    "Random Forest": RandomForestRegressor(n_estimators=100, random_state=42),
    "Gradient Boosting": GradientBoostingRegressor(n_estimators=100, random_state=42),
    "XGBoost": xgb.XGBRegressor(n_estimators=100, random_state=42, verbosity=0),
    "LightGBM": lgb.LGBMRegressor(n_estimators=100, random_state=42),
    "Support Vector Regressor": SVR()
}

for name, model in models.items():
    r2_scores = cross_val_score(model, X_encoded, y, cv=5, scoring='r2')
    neg_mse_scores = cross_val_score(model, X_encoded, y, cv=5, scoring='neg_mean_squared_error')
    rmse_scores = np.sqrt(-neg_mse_scores)

    print(f"{name}:")
    print(f"  Mean RÂ²: {r2_scores.mean():.4f} Â± {r2_scores.std():.4f}")
    print(f"  Mean RMSE: {rmse_scores.mean():.4f} Â± {rmse_scores.std():.4f}\n")
```
