```python
# Calculate missing values
missing_counts = df.isnull().sum()
features_withNaN = missing_counts[missing_counts > 0]

if not features_withNaN.empty:
    # Create a DataFrame with counts and percentages
    missing_data = pd.DataFrame({
        'Missing Count': features_withNaN,
        'Percentage (%)': (features_withNaN / len(df)) * 100
    }).sort_values(by='Missing Count', ascending=False)
    
    # Format the percentage to 2 decimal places
    missing_data['Percentage (%)'] = missing_data['Percentage (%)'].round(2)
    
    # Print the table
    print("Features with Missing Values:")
    print(missing_data)
else:
    print("No features with NaN values found.")
```

```python
# Drop columns with too many missing values more than 60 percent
df.drop(['Pool QC', 'Misc Feature', 'Alley', 'Fence'], axis=1, inplace=True)

# Replace FireplaceQu NaN with "No Fireplace"
df['Fireplace Qu'].fillna('No Fireplace', inplace=True)
```

```python
garage_cols = ['Garage Type', 'Garage Finish', 'Garage Qual', 'Garage Cond']
df[garage_cols] = df[garage_cols].fillna('No Garage')

# For Garage Yr Blt, impute with Year Built (since no garage = no year)
df['Garage Yr Blt'].fillna(df['Year Built'], inplace=True)
```

```python
bsmt_cat_cols = ['Bsmt Qual', 'Bsmt Cond', 'Bsmt Exposure', 'BsmtFin Type 1', 'BsmtFin Type 2']
df[bsmt_cat_cols] = df[bsmt_cat_cols].fillna('No Basement')
```

```python
df['Lot Frontage'] = df.groupby('Neighborhood')['Lot Frontage'].transform(
    lambda x: x.fillna(x.median())
)
```

```python
df['Mas Vnr Type'].fillna('None', inplace=True)
df['Mas Vnr Area'].fillna(0, inplace=True)
```

```python
df['Electrical'].fillna(df['Electrical'].mode()[0], inplace=True)
```

```python
bsmt_num_cols = ['BsmtFin SF 1', 'BsmtFin SF 2', 'Bsmt Unf SF', 'Total Bsmt SF', 'Bsmt Full Bath', 'Bsmt Half Bath']
df[bsmt_num_cols] = df[bsmt_num_cols].fillna(0)
```

```python
df.isnull().sum()
```

### Feature Engineering

```python
#Buyers care about total livable space, not just above-ground
df['TotalSF'] = df['Gr Liv Area'] + df['Total Bsmt SF']
```

```python
df.columns
```

```python
#summing all the bathrooms in the house
df['TotalBath'] = df['Full Bath'] + (0.5 * df['Half Bath']) + \
                  df['Bsmt Full Bath'] + (0.5 * df['Bsmt Half Bath'])
```

```python
#Age of the house and categoring into new or remodel
df['Age'] = df['Yr Sold'] - df['Year Built']
df['Remodeled'] = (df['Year Remod/Add'] != df['Year Built']).astype(int)
```

```python
#Total porch space
df['TotalPorchSF'] = df['Open Porch SF'] + df['Enclosed Porch'] + \
                     df['3Ssn Porch'] + df['Screen Porch']
```

```python
#if has basement or not
df['HasBasement'] = df['Total Bsmt SF'].apply(lambda x: 1 if x > 0 else 0)
```

```python
#if house has garage or not
df['HasGarage'] = df['Garage Area'].apply(lambda x: 1 if x > 0 else 0)
```

```python
#if the house has fireplace or not
df['HasFireplace'] = df['Fireplaces'].apply(lambda x: 1 if x > 0 else 0)
```

```python
df["Neighborhood"].unique()
```

```python
#neighborhood which are less relavent are grouped as others
neighborhood_counts = df['Neighborhood'].value_counts()
rare_neighborhoods = neighborhood_counts[neighborhood_counts < 10].index
df['Neighborhood_Grouped'] = df['Neighborhood'].replace(rare_neighborhoods, 'Other')
```

```python
df['Neighborhood_Grouped'].unique()
```

```python
#grouping house by style type 
df['HouseStyle_Simple'] = df['House Style'].replace({
    '1.5Unf': 'MultiLevel', '1.5Fin': 'MultiLevel',
    '2.5Unf': 'MultiLevel', '2.5Fin': 'MultiLevel',
    'SFoyer': 'SingleStory', 'SLvl': 'SingleStory',
    '1Story': 'SingleStory', '2Story': 'MultiLevel'
})
```

```python
df['HouseStyle_Simple'].unique()
```

```python
#High-quality materials in a large home may boost value
df['Qual_x_Area'] = df['Overall Qual'] * df['Gr Liv Area']
```

```python
#bathrooms per room
df['BathPerBedroom'] = df['TotalBath'] / df['Bedroom AbvGr']
df['BathPerBedroom'].replace(np.inf, 0, inplace=True)  
```

```python
df['Mo Sold'].unique()
```

```python
#converting months to quater
df['QuarterSold'] = df['Mo Sold'].replace({
    1: 1, 2: 1, 3: 1,      
    4: 2, 5: 2, 6: 2,    
    7: 3, 8: 3, 9: 3,     
    10: 4, 11: 4, 12: 4    
})
```

```python
#time since renovation
df['YearsSinceRemodel'] = df['Yr Sold'] - df['Year Remod/Add']
```

```python
#price per square foot
df['PricePerSF'] = df['SalePrice'] / df['TotalSF']
```

```python
df['Lot Shape'].unique()
```

```python
#numeric score for lot shape
df['LotShape_Score'] = df['Lot Shape'].map({
    'Reg': 3,   
    'IR1': 2,   
    'IR2': 1,  
    'IR3': 0    
})
```

```python
df.dropna(inplace=True)
```
